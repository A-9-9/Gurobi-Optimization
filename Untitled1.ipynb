```python
df = pd.DataFrame(
    [
        [date(2014, 1, 31), date(1999, 9, 24)],
        [datetime(1998, 5, 26, 23, 33, 4), datetime(2014, 2, 28, 13, 5, 13)],
    ],
    index=['x', 'y'],
    columns=['Date', 'Datetime'],
)
```

```python
df.head()
```

```python
with pd.ExcelWriter(
    'path_to_file.xlsx', 
    date_format="YYYY-MM-DD",
    datetime_format="YYYY-MM-DD HH:MM:SS"
) as writer:
    df.to_excel(writer, sheet_name='Sheet3')
```

```python
with pd.ExcelWriter(
    'path_to_file.xlsx', 
    mode='a', 
    engine="openpyxl", 
    if_sheet_exists="overlay",
) as writer:
    pd.DataFrame(
        [
            [1, 2],
            [3, 4]
        ],
        index=['x', 'y'],
        columns=['Date', 'Datetime'],
    ).to_excel(writer, sheet_name='Sheet3')
```
